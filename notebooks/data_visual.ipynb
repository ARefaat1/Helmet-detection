```python
import matplotlib.pyplot as plt
import os


#plot the images from the tarin dataset
train_images = [os.path.join("data", "train", f) for f in os.listdir("data/Motorcycle-Helmet-Detection-2/train")]
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    img = plt.imread(train_images[i])
    plt.imshow(img)
    plt.axis("off")
```
